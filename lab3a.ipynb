```python
!pip install gapminder
from gapminder import gapminder as df_gapminder
```

```python
df_gapminder.head()
```

**1.1 Find the most recent year for which we have data:**

```python
df_gapminder["year"].max()
```

**1.2 Create a new dataframe with data from 2007 only.**

```python
df_2007 = df_gapminder[df_gapminder["year"] == 2007]
df_2007.head()
```

**1.3 Create a scatterplot that graphs the relationship between a country's wealth per capita and its population's average life expectancy.**

```python
df_2007.plot.scatter(x="gdpPercap", y="lifeExp")
```

```python
df_2007["gdpPercap"].mean()
```

```python
df_2007["gdpPercap"].var() ** (1/2)
```

```python
df_2007["lifeExp"].mean()
```

```python
df_2007["lifeExp"].var() ** (1/2)
```

**In 2007, it appears that a country's wealth per capita relates to the log of its population's average life expectancy. As can be seen in the scatterplot, countries with a low GDP per capita ( < 10000) have a wide range of average life expectancies (40 - 75). On the other hand, countries with a medium to high GPD per capita ( > 10000) steadily have average life expectancies above 70. There are a few countries that don't follow the trend, mostly in the lower left (under where we could imagine a log curve).**

**2.1 Now using the original dataframe, find the average life expectancy for each year.**

```python
lifeExp_year = df_gapminder.groupby("year")["lifeExp"].mean()
lifeExp_year
```

**2.2 Create a line plot that shows how average life expectancy changed over time.**

```python
lifeExp_year.plot.line()
```

**Over time, life expectancy has increased almost linearly.**

```python
country_year_life = df_gapminder.groupby(["country", "year"])["lifeExp"].mean()
country_year_life
```

**There were instances where a country's life expectancy has fluctuated. I believe that some of these fluctuations occurred due to the events of the Cold War.**
